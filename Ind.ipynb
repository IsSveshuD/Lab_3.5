Постройте круговую диаграмму распределения суши на Земле по следующим данным:
    
| Вид суши	| Занимаемая площадь, млн.кв.км.	| Сектор диаграммы, градусы|
|:--:|:--:|:--:|
|Леса |	57 |--|	
|Степи|	24 |--|
|Тундры, пустыни и болота |	54	|--|
|Пашня|	15 |--|	
|Всего:|--| 360°|


```python
vals = np.array([57, 24, 54, 16])
valss = (vals/360) * 100

labels = ["Леса", "Степи", "Тундры, пустыни и болота", "Пашня"]
explode = (0.1, 0, 0.15, 0)

fig, ax = plt.subplots()

ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode,
wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, rotatelabels=True)
ax.axis("equal")
```
